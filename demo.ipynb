```python
import pandas as pd
from cleancore import CleanEngine
```

```python
df = pd.read_csv("Groceries_dataset.csv")
df.head()
```

```python
df.info()
df.shape
df.isnull().sum()
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower().str.strip()
df.columns
```

```python
df["itemdescription"].head()
```

```python
df = df.dropna(subset=["itemdescription"])
```

```python
df["itemdescription"] = (
    df["itemdescription"]
    .astype(str)
    .str.lower()
    .str.strip()
)
```

```python
df["date"] = pd.to_datetime(df["date"], errors="coerce")
```

```python
clean_df = df[["member_number", "date", "itemdescription"]]
clean_df.head()
```

```python
from cleancore import CleanEngine

engine = CleanEngine(df=clean_df)
cleaned_data, audit_log = engine.run()

cleaned_data.head()
audit_log
```

```python
before = len(df)

df = df.drop_duplicates()

after = len(df)
print("Rows before:", before)
print("Rows after:", after)
```

```python
df["itemdescription"] = (
    df["itemdescription"]
    .str.lower()
    .str.replace(r"[^a-z\s]", "", regex=True)
)
```

```python
transactions = (
    df.groupby(["member_number", "date"])["itemdescription"]
    .apply(list)
    .reset_index()
)

transactions.head()
```

```python
from cleancore.audit import AuditEvent


def run(self):
    self.data["itemDescription"] = (
        self.data["itemDescription"]
        .str.lower()
        .str.strip()
    )

    self.audit_log.append(
        AuditEvent("itemDescription lowercased and stripped")
    )

    return self.data, self.audit_log
```

```python
engine = CleanEngine(df)
clean_df, audit = engine.run()

clean_df["itemDescription"].head()
audit
```
